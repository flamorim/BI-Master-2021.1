```
# exercicio 01

#variaveis
resultado = False

#funcoes
def soma(a,b):
   return a + b

def subtrai(a,b):
   return a - b

def divide(a,b):
   return a / b

def multiplica(a,b):
   return a * b

#Codigo principal2
num_1 = float(input("Favor entrar com o primeiro número: "))
num_2 = float(input("favor entrar com o segundo número: "))
#ope = input("Favor informar a operação desejada: ")

ope = None 
while ope not in {"soma", "divisão","multiplicação","subtração","q"}: 
         ope = input("Favor informar a operação desejada. Responda com 'soma', 'divisão','multiplicação','subtração' ou 'q' para sair: ")
         
if ope.lower() == "soma":
  resultado = soma(num_1,num_2)
elif ope.lower() == "divisão":
  resultado = divide(num_1,num_2)
elif ope.lower() == "multiplicação":
  resultado = multiplica(num_1,num_2)
elif ope.lower() == ("subtração"):
  resultado = subtrai(num_1,num_2)

if ope.lower() == ("q"):
  print("programa abortado")
elif resultado:
  print(f'O resultado é: {resultado}')
  # par ou impar somente para numeros inteiros  
  if (resultado // 1) == resultado:
    print("É um número inteiro")
    if (resultado % 2):
      print("É um número impar")
    else:
      print("É um número par")
  else:
    print("É um número decimal")
  if ( resultado < 0):
    print("É um número negativo")
  else:
    print("É um número positivo")

```

```
#exercicio 5
lista1 = []
lista2 = []
lista3 = []
lista4 = []
entrada = []
cont1 = 0
cont2 = 0
cont3 = 0
cont4 = 0


inicio = 0
fim = 25
cont = inicio
while cont <= fim:
  lista1.append(cont)
  cont +=1

inicio = 26
fim = 50
cont = inicio
while cont <= fim:
  lista2.append(cont)
  cont +=1

inicio = 51
fim = 75
cont = inicio
while cont <= fim:
  lista3.append(cont)
  cont +=1

inicio = 76
fim = 100
cont = inicio
while cont <= fim:
  lista4.append(cont)
  cont +=1


flag = False
while flag == False:
  try:
    temp = input("Entre com um numero: ")
    num = int(temp)
  except ValueError:
    print ("Entrada Invalida")
  else:
    if (num < 0):
      flag = True
    else:
      entrada.append(num)

fim = len(entrada)
cont = 0

while cont < fim:
  if entrada[cont] in lista1:
    cont1 += 1
  if entrada[cont] in lista2:
    cont2 += 1
  if entrada[cont] in lista3:
    cont3 += 1
  if entrada[cont] in lista4:
    cont4 += 1
  cont += 1
print(f'No intervalo1: {cont1}')
print(f'No intervalo2: {cont2}')
print(f'No intervalo3: {cont3}')
print(f'No 1intervalo4: {cont4}')

```
